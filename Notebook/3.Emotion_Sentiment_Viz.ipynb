```python
bot = ['NOWARNONATONOU1','Erica43581765', 'PEACEFREEDOM20256268', 'savedem65052841', 'Eric57017912']
```

```python
df_emot = pd.read_csv('###')
df_hub_final=df_emot[~df_emot['author_name'].isin(bot)].reset_index(drop=True)
```

```python
df_hub_final.head()
```

# SENTIMENT

```python
## CALCOLO sentiment su questi testi
df_hub_final['sentiment'] = df_hub_final['sentiment_text'].apply(lambda x: Analyzer.polarity_scores(str(x)))
## CREAZIONE colonne neg, neu, pos, compound
df_hub_final['neg']=df_hub_final['sentiment'].apply(lambda x: x['neg'])
df_hub_final['neu']=df_hub_final['sentiment'].apply(lambda x: x['neu'])
df_hub_final['pos']=df_hub_final['sentiment'].apply(lambda x: x['pos'])
df_hub_final['compound']=df_hub_final['sentiment'].apply(lambda x: x['compound'])
## CREAZIONE COLONNA NEW_SENT BASATO SU VALORE COMPOUND
new_sent = []
for i,x in enumerate(df_hub_final.iterrows()):
  comp = df_hub_final.loc[i,'compound']
  if 0.05 > comp > -0.05:
    new_sent.append('neutrale')
  if comp > 0.05:
    new_sent.append('positivo')
  if comp < -0.05:
    new_sent.append('negativo')

df_hub_final['new_sent']=new_sent
```

```python
## salvataggio
df_hub_final.to_csv('###', index=False)
```

```python
polarity_intra_cluster = df_hub_final.groupby(["cluster","new_sent"]).count()["id"].unstack(fill_value=0)
polarity_intra_cluster.head()
```

```python
emotions_intra_cluster = df_hub_final.groupby(["cluster","emotion"]).count()["id"].unstack(fill_value=0)
emotions_intra_cluster.head()
```

```python
df_cluster = polarity_intra_cluster.join(emotions_intra_cluster, how = "right")
# df_cluster.drop(index = -1, inplace = True) # elimino il cluster -1 (già fatto)
df_cluster["cluster_size"] = df_hub_final.groupby("cluster").size() 
```

```python
df_cluster["positivo"] = df_cluster["positivo"] / df_cluster["cluster_size"] *100
df_cluster["negativo"] = df_cluster["negativo"] / df_cluster["cluster_size"] *100
df_cluster["neutrale"] = df_cluster["neutrale"] / df_cluster["cluster_size"] *100

df_cluster["fear"] = df_cluster["fear"] / df_cluster["cluster_size"] *100
df_cluster["anger"] = df_cluster["anger"] / df_cluster["cluster_size"] *100
df_cluster["sadness"] = df_cluster["sadness"] / df_cluster["cluster_size"] *100
df_cluster["joy"] = df_cluster["joy"] / df_cluster["cluster_size"] *100


df_cluster = df_cluster.round(2)
df_cluster.head(10)
```

# WORDCLOUD

```python
df_politici_final = pd.read_csv('###')
df_politici_final.head()
```

```python
x = df_politici_final.groupby(['cluster'], as_index = False)['id'].count()
lista_conteggio = x['id']
```

```python
# dataset finale con cluster raggruppati e corpus dei testi uniti
word_final = df_politici_final.groupby(['cluster'], as_index = False).agg({'text_clean': list})
word_final['numerosità'] = lista_conteggio
# ordino per numerosità dei gruppi, per capire quante wordcloud mostrare
word_final = word_final.sort_values(by=['numerosità'], ascending=False)
```

```python
word_final
```

```python
for i in range (0, len(word_final)):
  word_final['text_clean'][i] = ''.join(str(x) for x in word_final['text_clean'][i])
```

```python
sns.set()
plt.rcParams['figure.figsize'] = [25, 50]
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.4, wspace=0.2, hspace=0.1)

wordcloud = WordCloud(background_color="white", contour_width=0.1, 
                      contour_color="black",  max_font_size=150, random_state=42,
                      colormap="Dark2", stopwords=stop)

for i in range(5):
    wordcloud.generate(text=word_final["text_clean"][i])   
    plt.subplot(5, 5, i+1)
    plt.imshow(wordcloud, interpolation="bilinear")
    tit = 'Cluster: ' + str(word_final["cluster"][i])
    plt.title(tit)
    plt.axis("off")
plt.show()
```
